```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
ages = df.groupby("age").median().reset_index()
print(df.head())
print(ages)
df.info()
```

```python
ages.plot.line("age", "chol");
plt.show()
```

```python
# matplotlib 
# ls = linestyle
# lw = line width 
fig, axes = plt.subplots()
axes.plot(ages['age'], ages['chol'], ls = '-', lw = 0.7)
axes.set_xlabel("Age")
axes.set_ylabel("Cholestrol")
plt.show()
```

```python
# using seaborn
sns.lineplot(x = "age", y = "chol", data = ages, linestyle = ":", linewidth = 1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

```python
bp = df.groupby("trestbps").median().reset_index()
bp.plot.line("age", "trestbps")
plt.show()
```

#### pd.offsets.BDay()

- Resampling: You can use BDay() when resampling time series data by business days.
- Custom Offsets: You can combine BDay() with numbers to shift dates by multiple business days.

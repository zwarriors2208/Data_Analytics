```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(1,10,10)
b = a ** 2
```

```python
x = np.arange(0,10)
y = 2*x
```

- fig.add_axes([left, bottom, width(right), height])
- - add_axes(): This is used when you need fine-grained control over the exact placement of the axes (like creating insets, overlapping plots, or irregular layouts).

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

## sub plots

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(12,8))
axes[0][0].plot(a,b)
axes[0][1].plot(x,y)
axes[1][0].plot(y,x)
axes[1][1].plot(b,a)
# use left, right, top, bottom to stretch the subplots
# use wspace, hspace to add spacing between subplots - wspace is vertical space between subplots - hspace is horizontal space between 
fig.subplots_adjust(left = 0,
                    right = 1,
                    bottom = 0,
                    top = 1,
                    wspace = 0.5,
                    hspace=0.5)
# fig.subplots_adjust(left = None,
#                     bottom = None,
#                     top = None,
#                     wspace = 0.5,
#                     hspace=0.5)
plt.show()
```

- indexing is done using grids  plt.subplot(row=2, column=1, index=1)

```python
fig = plt.figure()

plt.subplot(2, 1, 1)
plt.xticks(())
plt.yticks(())

plt.subplot(2, 3, 4)
plt.xticks(())
plt.yticks(())

plt.subplot(2, 3, 5)
plt.xticks(())
plt.yticks(())

plt.subplot(2, 3, 6)
plt.xticks(())
plt.yticks(())

plt.show()
```

## scatter and other plots

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
plt.figure(figsize = (8,6))
plt.scatter(x, y, color = 'blue', alpha = 0.5)
plt.title("scatter plot using matplotlib", fontsize = 15)
plt.xlabel('X-axis', fontsize = 12)
plt.ylabel('Y-axis', fontsize = 12)
plt.grid(True)
plt.show()
```

- creating same plot using sns

```python
import seaborn as sns
sns.set(style = 'whitegrid') # selecting style with white grid background
plt.figure(figsize = (8, 6))
sns.scatterplot(x = x, y=y)
plt.title('Scatter plot using seaborn', fontsize = 15)
plt.show()
```

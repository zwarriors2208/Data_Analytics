```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
sns.set(style = 'darkgrid') # selecting style with white grid background
plt.figure(figsize = (8, 6))
sns.scatterplot(x = x, y=y, alpha = 0.9, c = 'r')
plt.title('Scatter plot using seaborn', fontsize = 15)
plt.show()
```

### line plot

```python
import pandas as pd
df = pd.read_csv("heart.csv")
ages = df.groupby("age").median().reset_index()
print(df.head())
df.info()
```

```python
sns.lineplot(x = "age", y = "chol", data = ages, linestyle = ":", linewidth = 1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

## histogram(count plots)

```python
df = pd.read_csv('dm_office_sales.csv')
## Countplot: A simple plot, it merely shows the total count of rows per category.
# A histogram across a categorical, instead of quantitative, variable

plt.figure(figsize=(10,4),dpi=200) # dpi=Dots Per Inch of space
sns.countplot(x='division',data=df)
plt.show()

plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education', data=df)
plt.show()
```

- The hue parameter allows you to split the data by a categorical variable and assign different colors to different categories.

```python
#Breakdown within another category with 'hue'
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education', data=df, hue='training level')
plt.show()
```

```python
#Using palette
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df,hue='training level', palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(10,4),dpi=200)
#Paired would be a good choice if there was a distinct jump from 0 and 1 to 2 and 3 
sns.countplot(x='level of education',data=df,hue='training level',palette='Paired')
plt.show()
```

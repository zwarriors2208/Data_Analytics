```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df['age'] = df['age'].fillna(int(df['age'].mean()))
```

```python
df['deck'] = df['deck'].fillna(df['deck'].mode()[0])
```

```python
df.info()
```

```python
df['sex'] = df['sex'].astype('str')
```

```python
df.info()
```

```python
df
```

```python
df.groupby('sex')['survived'].sum().plot(kind = 'bar')
```

```python
df.groupby('class')['fare'].mean().plot()
```

```python
df.groupby('class')['survived'].sum().plot(kind = 'bar')
```

```python

df.plot(x = 'alive', y = 'age', kind = 'bar')
```

```python
df.info()
```

```python
df['alive'] = df['alive'].astype('category')
```

```python
df.plot(x = 'alive', y = 'age', kind = 'scatter')
```

```python
df
```

```python
df.groupby(['deck', 'class'])['survived'].sum().plot(kind = 'bar')
```

```python
df.groupby(['deck', 'class'])['fare'].mean().plot(kind = 'bar')
```

```python
df.groupby('alone')['survived'].sum().plot(kind = 'bar')
```

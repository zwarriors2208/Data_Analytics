#### 1. Write a Python programming to display a bar chart of the popularity of programming Languages.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Programming_languages=['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
plt.bar(Programming_languages,Popularity)
plt.title('popularity of programming Languages.')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.minorticks_on()
plt.grid(which = 'major', linestyle = '--')
plt.grid(which='minor', ls=':', lw='0.5', c='red', alpha = 0.2)
```

#### 10.Write a Python program to create bar plot of scores by group and gender. 
Use multiple X values on the same chart for men and women.

```python
men = [22, 30, 35, 30, 26]
women = [25, 32, 30, 35, 29]
g = np.array([0,1,2,3,4])
plt.bar(g,men, width = 0.3, label = 'MEN')
plt.bar(g + 0.3,women, width = 0.3, label = 'WOMEN')
plt.margins(x = 0)
G = ['G1', 'G2', 'G3', 'G4', 'G5']
plt.xticks(g+0.15, G)
plt.legend()
plt.xlim(right = 4.6)
plt.xlabel("Person")
plt.ylabel("Scores")
plt.title("Scores By Person")
plt.show()
```

#### 12. Write a Python program to create bar plots with error bars on the same figure.

```python
mv = np.array([0.2474, 0.1235, 0.1737, 0.1824])
sd = np.array([0.03314, 0.02278, 0.02836, 0.02645])
x = [1,2,3,4]
```

```python
plt.bar(x, mv, width = 0.5, yerr = sd, color = 'r', edgecolor = 'black')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green', alpha = 0.2)
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black', alpha = 0.2)
plt.xlabel('velocity')
plt.ylabel('score')
plt.title('Scores by velocity')
```

#### 13. Write a Python program to create bar plots with errorbars on the samefigure. Attach a text label above each bar displaying men means (integervalue).

```python
plt.bar(x, mv, width = 0.5, yerr = sd, color = 'r', edgecolor = 'black', label = "Men")
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green', alpha = 0.2)
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black', alpha = 0.2)
plt.xlabel('velocity')
plt.ylabel('score')
plt.title('Scores by velocity')
plt.legend()
for i in range(4):
    plt.text(i+1, mv[i] + sd[i], f'{mv[i] : .2f}', ha='center', va='bottom')
```

#### 14. Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.

```python
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
inx=np.arange(5)
```

```python
plt.bar(inx, menMeans,width=0.5, yerr=menStd,color='r',edgecolor='black',label='men',capsize=5)
plt.bar(inx, womenMeans,bottom=menMeans,width=0.5, edgecolor='black', yerr=womenStd,color='green',label='women',capsize=5)
plt.legend()
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.xticks([0,1,2,3,4],['Group1','Group2','Group3','Group4','Group5'])
plt.title('Scores by group and gender')
```

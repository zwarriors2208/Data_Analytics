#### 3. Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.

```python
start_date = '2020-04-01'
end_date = '2020-04-07'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
filtered_df.plot(kind = 'bar',x = 'Date', y = 'Volume')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation = 45)
plt.show()
```

#### 14. Write a Pandas program to create a plot of adjusted closing prices, thirty dates.days and forty days simple moving average of Alphabet Inc. between two specific

```python
df3 = df1
close_px = df3['Adj Close']
df3['SMA_30_days'] = df3.iloc[:,4].rolling(window=30).mean() 
df3['SMA_40_days'] = df3.iloc[:,4].rolling(window=40).mean()
plt.figure(figsize=[10,8])
plt.grid(True)
plt.title('Historical stock prices of Alphabet Inc. [01-04-2020 to 30-09-2020]\n',fontsize=18, color='black')
plt.plot(df3['Adj Close'],label='Adjusted Closing Price', color='black')
plt.plot(df3['SMA_30_days'],label='30 days simple moving average', color='red')
plt.plot(df3['SMA_40_days'],label='40 days simple moving average', color='green')
plt.legend(loc=2)
plt.show()
```

#### 18. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates.

```python
df2 = df1[['Date', 'Close']].set_index('Date')
d = df2.pct_change()
d = df2.rolling(window = 30, min_periods = 30).std()
plt.figure(figsize=(20,20))
d.plot();
plt.grid(True)
plt.show() 
```

```python
df2 = df1[['Date', 'Close']]
df3 = df2.set_index('Date')
data_filled = df3.asfreq('D', method='ffill')
data_returns = data_filled.pct_change()
data_std = data_returns.rolling(window=30, min_periods=30).std()
plt.figure(figsize=(20,20))
data_std.plot();
plt.suptitle('Volatility over a period of time  of Alphabet Inc. stock price,\n01-04-2020 to 30-09-2020', fontsize=12, color='black')
plt.grid(True)
plt.show() 
```

```python
from pulp import *

# Create the problem variable to contain the prov=blem data
model = LpProblem('CarProductionProblem', LpMaximize)

# Create 2 variables for Cars A and B
# Parameters: Name, Lower limit, Upper limit, Data type
car_A = LpVariable("A", 0, None, LpInteger) # Car A prdouction
car_B = LpVariable("B", 0, None, LpInteger) # Car B production

# Create maximize objective function(pofit)
model += 30000 * car_A+ 45000 * car_B, "Profit"

#Create three constraints
model += 3* car_A + 4 * car_B <= 30,"Material Constraint 1" # Constraint 1
model += 5* car_A + 6 * car_B <= 60, "Material Constraint 2" # Constraint 2
model += 1.5* car_A + 3 * car_B  <= 21, "Labour Constraint" # Constraint 3

#Optional constraint - Minimum production of Car A (commented out)
# model += car_A >= 5, 'Minimum A Production' 

#The problem is solved using PULP's choice of Solver
model.solve()

#Each of the variables is printed with it's resolved optimum value
for v in model.variables():
    print(v.name, '=', v.varValue)

# Additional Information  (Optional)
print("Total Profit:", value(model.objective)) # Print objective function value (Total Profit)
```
